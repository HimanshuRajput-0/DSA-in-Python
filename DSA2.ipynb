**Python Collections (Arrays) :**
```
There are four collection data types in the Python programming language:

List is a collection which is ordered and changeable. Allows duplicate members.
Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.
Dictionary is a collection which is ordered** and changeable. No duplicate members.
```

**List :**```Lists are used to store multiple items in a single variable.```

```Lists are created using square brackets.```

```Ex :```

````python
list = [20,25,30,"himanshu"]
````

**List items :**```List items are ordered, changeable, and allow duplicate values.```

```
When we say that lists are ordered, it means that the items have a defined order, and that order will not change.

If you add new items to a list, the new items will be placed at the end of the list.
```

```The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.```

```List items are indexed, the first item has index [0], the second item has index [1].```

```Since lists are indexed, lists can have items with the same value.```

```List items can be of any data type.```

```A list can contain different data types.```

**List lenght :**``` To determine how many items a list has, use the len() function.```

```Ex :```

````python
list = [20,25,30,"himanshu"]
print(len(list))
````

**Access List items :**``` List items are indexed and you can access them by referring to the index number.```

```The first item has index 0.```

```Ex :```

````python
list = ['a', 'b', 'c']
print(list[1]) # accessing second element
````

**Negative Indexing :**```Negative indexing means start from the end.```

```-1 refers to the last item, -2 refers to the second last item etc.```

```Ex :```

````python
list = ['a', 'b', 'c']
print(list[-1]) # accessing last element
````

**Range of Indexes :**``` We can specify a range of indexes by specifying where to start and where to end the range.```

```When specifying a range, the return value will be a new list with the specified items.```

```Note: The search will start at index 2 (included) and end at index 5 (not included).```

```Ex :```

````python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
````

```By leaving out the start value, the range will start at the first item.```

```Ex :```

````python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[:5])
````

```By leaving out the end value, the range will go on to the end of the list.```

```Ex :```

````python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:])
````

**Range of Negative Indexes :**

```Ex :```

````python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[-4:-1])
````

```Check if Item Exists : To determine if a specified item is present in a list use the in keyword.```

```Ex :```

````python
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes")
````

**Change list items :**```To change the value of a specific item, refer to the index number.```

```Ex :```

````python
list = ['a', 'b', 'c']
list[1] = 'd'
print(list)
````

```To change the value of items within a specific range, define a list with the new values, and refer to the range of index numbers where we want to insert the new values.```

```Ex :```

````python
list = ['a', 'b', 'c', 'd', 'e']
list[1:3] = ['f', 'g']
print(list)
````

```If we insert more items than we replace, the new items will be inserted where we specified, and the remaining items will move accordingly.```

```Ex :```

````python
list = ['a', 'b', 'c', 'd', 'e']
list[1:2] = ['f', 'g']
print(list)
````

```If we insert less items than we replace, the new items will be inserted where we specified, and the remaining items will move accordingly.```

```Ex :```

````python
list = ['a', 'b', 'c', 'd', 'e']
list[1:3] = ['f']
print(list)
````

**Inser items :**``` To insert a new list item, without replacing any of the existing values, we can use the insert() method.```

```The insert() method inserts an item at the specified index.```

```Ex :```

````python
list = ['a', 'b', 'c', 'd', 'e']
list.insert(2, 'f')
print(list)
````
